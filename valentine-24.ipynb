```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
```

```python
import os
df = pd.read_csv('/kaggle/input/valentine24/Valentine 24 Analysis - Analysis.csv')
```

```python
# Sort by quantity sold and get the top 10 products
top_10_quantity = df.sort_values(by='Quantity', ascending=False).head(10)

# Create a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Product', y='Quantity', data=top_10_quantity, palette='viridis')
plt.title('Top 10 Products by Quantity')
plt.xlabel('Product')
plt.ylabel('Quantity')
plt.xticks(rotation=45)
plt.show()
```

```python
# Strip spaces from column names
df.columns = df.columns.str.strip()

# Sort by Revenue and get the top 10 products
top_10_revenue = df.sort_values(by='Total Revenue', ascending=False).head(10)

# Create a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Product', y='Total Revenue', data=top_10_revenue, palette='magma')
plt.title('Top 10 Products by Total Revenue Generated')
plt.xlabel('Product')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Quantity', y='Total Revenue', data=df, hue='Product', palette='coolwarm', s=100)
plt.title('Total Revenue vs. Quantity')
plt.xlabel('Quantity')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Move legend outside the plot
plt.show()
```
